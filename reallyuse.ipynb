```python
import torch
from torchvision import models, transforms
from PIL import Image

# 创建模型架构
model = models.resnet50(pretrained=False)  # 用 pretrained=False 确保不加载预训练权重
num_ftrs = model.fc.in_features
model.fc = torch.nn.Linear(num_ftrs, 2)  # 与之前训练时相同的类别数

# 加载模型权重
model.load_state_dict(torch.load('model_resnet50.pth'))
model.eval()  # 设置为评估模式
```

```python
# 定义图像预处理
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# 加载图像
img = Image.open(r'F:\AI\resnet50\data\test\269.jpg')  # 修改为你的图像路径
img_t = transform(img)
batch_t = torch.unsqueeze(img_t, 0)
```

```python
# 使用模型进行预测
with torch.no_grad():
    out = model(batch_t)

# 获取最可能的类别
_, predicted = torch.max(out, 1)
print(f'Predicted class index: {predicted.item()}')
```

```python
# 假设你有一个类别索引到名称的映射
classes = ['bing', 'zu']  # 根据你的类别填充这个列表
predicted_class = classes[predicted.item()]
print(f'Predicted class name: {predicted_class}')
```
